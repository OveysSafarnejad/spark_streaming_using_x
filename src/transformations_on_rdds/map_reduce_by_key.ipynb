```python
import findspark

findspark.init('/opt/spark-3.5.0/')

from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext

spark = SparkSession.builder.appName('rdd transformations').getOrCreate()
sc = spark.sparkContext
ssc = StreamingContext(sc, 1)
```

```python
sc.parallelize([j for j in range(1, 25)])
```

```python
import time


rddQueue = []
for i in range(5):
    rddQueue += [sc.parallelize([j for j in range(1, 25)])] # 10 is the number of slices
```

# reduceByKey

```python
inputStream = ssc.queueStream(rddQueue)
mappedStream = inputStream.map(lambda x: (x % 10, 1))
reducedStream = mappedStream.reduceByKey(lambda a, b: a + b)
reducedStream.pprint()

ssc.start()
```

```python
time.sleep(6)
ssc.stop(stopSparkContext=True, stopGraceFully=True)
```
